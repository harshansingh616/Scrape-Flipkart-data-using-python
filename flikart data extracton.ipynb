**Scrape Flipkart data using python, beautifulsoup tutorial**

```python
#importing neccessary libraries  
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#Copy flipkart url from your local browser
URL = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

Then we need to define HTTP header, so when we visit any website on the internet we send it HTTP request. HTTP request contains a lot of info and one of them is header,header also contains lot of info and one of them is "USER AGENT". USER AGENT tells them that you are trying to access their website and you ae a genuine user by identifying your browser information and some of the other information required to access the website.

**Step to get "USER AGENT"**
1. Go to "whatismybrower" here's link " https://www.whatismybrowser.com/ "

2. Then click "Detect My Settings"

```python
# Header for request
HEADER= ({ 'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36" ,'Accept-Language': 'en-US,en;q=0.5' })
```

```python
#HTTP request
WebPage=requests.get(URL,headers=HEADER)
```

```python
#checking response
WebPage
```

```python
#object soup for collecting the data
Soup =  BeautifulSoup(WebPage.content,"html.parser")
Soup
```

```python
#lets extract the link from our "Webpage"
Link = Soup.find_all("a", attrs={'class': '_1fQZEK'})
print(Link)
```

```python
link=Link[0].get('href')
link
```

```python
Product_List="https://flipkart.com" + link
Product_List
```

```python
Product_WebPage=requests.get(Product_List,headers=HEADER)
Soup1= BeautifulSoup(Product_WebPage.content,"html.parser")
Soup1
```

```python
Title1 = Soup1.find("span", attrs={'class': 'B_NuCI'}).text
Title1
```

```python
Discount_Price1 = Soup1.find("div", attrs={'class': '_30jeq3 _16Jk6d'}).text
Discount_Price1
```

```python
Original_Price1 = Soup1.find("div", attrs={'class': '_3I9_wc _2p6lqe'}).text
Original_Price1
```
