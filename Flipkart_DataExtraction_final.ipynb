```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_title(soup):
    try:
        Title = soup.find("span", attrs={'class': 'B_NuCI'}).text
        
    except AttributeError:
        Titile=""
    return Title

def get_discount_price(soup):
    try:
        Discount_Price = soup.find("div", attrs={'class': '_30jeq3 _16Jk6d'}).text
        Discount_Price
    except AttributeError:
        Discounted_price = ""
    return Discount_Price

def get_original_price(soup):
    try:
        Original_Price = soup.find("div", attrs={'class': '_3I9_wc _2p6lqe'}).text
        Original_Price
    except AttributeError:
        Original_Price=""
    return Original_Price

def get_rating(soup):
    try:
        Rating=soup.find("div", attrs={'class': '_3LWZlK'}).text
    except AttributeError:
        Rating=""
    return Rating

def get_box_item(soup):
    try:
        Box_Item=soup.find('li',attrs={'class':'_21lJbe'}).text
    except AttributeError:
        Box_Item=""
    return Box_Item
        
        
```

```python
if __name__ == '__main__':
    URL = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
    
    HEADER= ({ 'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36" ,'Accept-Language': 'en-US,en;q=0.5' })
    
    WebPage=requests.get(URL,headers=HEADER)
    
    Soup =  BeautifulSoup(WebPage.content,"html.parser")
    
    Links = Soup.find_all("a", attrs={'class': '_1fQZEK'})
    #_1fQZEK
    print(type(Links))
    
    
    print(Links[0])
    print(Links[1])
    print(Links[2])
    print(Links[3])
    print(Links[4])
    
    Link_List=[]
    print(type(Link_List))
    print(len(Links))
    
    for link in Links:
        Link_List.append(link.get('href'))
        print(Link_List)
        
    #print(len(Link_List))
    #for i in Link_List:
        #print(Link_List[0])
        
    dict = {"title":[], "discounted_price":[],"original_price":[], "rating":[], "box_item":[]}
    
    for link in Link_List:
        new_webpage = requests.get("https://www.flipkart.com" + link, headers=HEADER)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")
    
    
    dict['title'].append(get_title(new_soup))
    dict['discounted_price'].append(get_discount_price(new_soup))
    dict['original_price'].append(get_original_price(new_soup))
    dict['rating'].append(get_rating(new_soup))
    dict['box_item'].append(get_box_item(new_soup))
    
    flipkart_df = pd.DataFrame.from_dict(dict)
    flipkart_df['title'].replace('', np.nan, inplace=True)
    flipkart_df = flipkart_df.dropna(subset=['title'])
    flipkart_df.to_csv("flipkart_data.csv", header=True, index=False)
    
```
